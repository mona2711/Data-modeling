```python
import pandas as pd
import nltk

#paras to sentences
from nltk.tokenize import sent_tokenize

#sentence to words
from nltk.tokenize import TreebankWordTokenizer

#stopwords
from nltk.corpus import stopwords

import string

# regular expression
import re

#emoji's
import demoji
```

```python
%%time 
#read IRAhandle_tweets_1.csv
file = pd.read_csv('IRAhandle_tweets_1.csv')
len(file)
```

```python
# fetching content column of file into csv without header 
text = file.to_csv(columns=['content'], index=False, header=False)
```

```python
# text[:1000]
```

```python
# # removing hyperlinks
print(len(text))

text = re.sub(r'(https?:\/\/)(\s)*(www\.)?(\s)*((\w|\s)+\.)*([\w\-\s]+\/)*([\w\-]+)((\?)?[\w\s]*=\s*[\w\%&]*)*', '', text, flags=re.MULTILINE)
print(len(text))

text = re.sub(r'\'\'', '', text, flags=re.MULTILINE)
print(len(text))

text = re.sub(r'\"\"', '', text, flags=re.MULTILINE)
print(len(text))
```

```python
# text[:1000]
```

```python
%%time

#converting sentences(tweets) into list of words
l_words = TreebankWordTokenizer().tokenize(text)
```

```python
%%time

#removing punctuations from words list
p = string.punctuation
l_words_p = [w for w in l_words if w not in p]
```

```python
l_words_p.sort()
# type(text)
```

```python
%%time

#emojis in corpus

emoji_set = demoji.findall(text)

l = sent_tokenize(text)
print(len(l))
```

```python
# emojis, links are removed

l_words_p1 = l_words_p[l_words_p.index('a'):]
```

```python
#eliminating stop-words

sw = stopwords.words('english')

l_words_sw = [w for w in l_words_p1 if w not in sw]
```
