```python
temp = []
df_col = [] 
l = len(file_1)
c = 0

for i in range(l):
    temp.append(file_1['user1_Tags'][i])
    temp.append(file_1['user2_Tags'][i])
    vect = TfidfVectorizer(min_df=1)
    tfidf = vect.fit_transform(temp)
    pairwise_similarity = tfidf * tfidf.T   
#     pairwise_similarity[0,0] = 0
#     pairwise_similarity[1,1] = 0
    if pairwise_similarity[0,1] > 0 : 
        c += 1
        df_col.append('similar users based on aboutme and tags')
    else :
        df_col.append('users are similar based on aboutme but not on tags')
    temp.clear()
```

```python
file_1['new_col'] = df_col
```

```python
file_2 = file_1[file_1['new_col']=='similar users based on aboutme and tags']
```
